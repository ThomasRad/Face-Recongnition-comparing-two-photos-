```python
import cv2
from google.colab.patches import cv2_imshow
import face_recognition
```

```python
img = cv2.imread("thomas.jfif")
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_encoding = face_recognition.face_encodings(rgb_img)[0]
cv2_imshow(img)
```

```python
img2 = cv2.imread("thomas2.jpg")
rgb_img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img_encoding2 = face_recognition.face_encodings(rgb_img2)[0]
cv2_imshow(img2)
```

```python
img3 = cv2.imread("luke.jfif")
rgb_img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
img_encoding3 = face_recognition.face_encodings(rgb_img3)[0]
cv2_imshow(img3)
```

```python
# Is Thoams = Thomas ?  

result = face_recognition.compare_faces([img_encoding],img_encoding2)
print(result)
```

```python
# Is Thomas = Jedi Knight ? 

result = face_recognition.compare_faces([img_encoding],img_encoding3)
print(result)
```
